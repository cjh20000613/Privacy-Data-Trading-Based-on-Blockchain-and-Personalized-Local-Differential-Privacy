```python
import pandas as pd
import numpy as np
import glob

folder_path = 'E:/anacoda_workspace/论文/新建文件夹 (2)/数据一—肥胖年龄/'  # 替换为实际的文件夹路径
file_pattern = 'result_12_experiment_*.xlsx'

# 获取符合规则的文件列表
file_list = glob.glob(f'{folder_path}{file_pattern}')
```

```python
# 读取并合并文件
dfs = []
for file in file_list:
    df = pd.read_excel(file)
    dfs.append(df)

# 合并所有数据框
combined_df = pd.concat(dfs, ignore_index=True)

# 打印合并后的数据框
print(combined_df)
```

```python
test_df = combined_df
test_df
```

```python
selected_data_list = []

for i in range(0, 20):
    selected_rows = list(range(i, 981 + i, 20))
    selected_data = test_df.iloc[selected_rows]
    selected_data_list.append(selected_data)
```

```python
# 使用列表推导和 mean 函数分别计算每个数据框的平均值
averages_list = [df.mean() for df in selected_data_list]
averages_list
```

```python
# 使用 pd.concat 将平均值列表重新组合成一个数据框
result_df = pd.concat(averages_list, axis=1)
result_df_transposed = result_df.T
result_df_transposed
```

```python
import matplotlib.pyplot as plt
# 实验一

# 选择要绘制的列
columns_to_plot = ['RE_values_1', 'RE_LAP_values_1']

# 预先设定不同列名对应的标签
column_labels = {'RE_values_1': 'BPPDT', 'RE_LAP_values_1': 'LAP'}

# 获取列名作为 x 轴数据
x_values = np.linspace(0, 1, len(result_df_transposed.index))

# 获取每列的平均值作为 y 轴数据，并乘以 100
for column in columns_to_plot:
    y_values = result_df_transposed[column] * 100
    
    # 绘制曲线
    plt.plot(x_values, y_values, label=column_labels.get(column, column))

# 设置 x 轴范围为 [0, 1]
plt.xlim(0, 1)

# 设置 x 轴标签
plt.xlabel('B_prime')

# 设置 y 轴标签
plt.ylabel('RE(%)')

# 添加图例
plt.legend()

# 显示图形
plt.show()
```
