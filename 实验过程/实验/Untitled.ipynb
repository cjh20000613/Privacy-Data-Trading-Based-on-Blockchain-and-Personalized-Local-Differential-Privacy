```python
folder_path = 'E:/anacoda_workspace/论文/新建文件夹 (2)/数据四—强迫症持续时间/'  # 替换为实际的文件夹路径
file_pattern = 'result_12_experiment_*.xlsx'

# 获取符合规则的文件列表
file_list = glob.glob(f'{folder_path}{file_pattern}')
# 读取并合并文件
dfs = []
for file in file_list:
    df = pd.read_excel(file)
    dfs.append(df)

# 合并所有数据框
combined_df = pd.concat(dfs, ignore_index=True)
test_df = combined_df
selected_data_list = []

for i in range(0, 20):
    selected_rows = list(range(i, 981 + i, 20))
    selected_data = test_df.iloc[selected_rows]
    selected_data_list.append(selected_data)

# 使用列表推导和 mean 函数分别计算每个数据框的平均值
averages_list = [df.mean() for df in selected_data_list]
# 使用 pd.concat 将平均值列表重新组合成一个数据框
result_df = pd.concat(averages_list, axis=1)
data4_df = result_df.T
data4_df
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D

# 选择要绘制的列
columns_to_plot = ['RE_values_1', 'RE_LAP_values_1']

# 预先设定不同列名对应的标签和颜色
column_info = {'RE_values_1': {'label': 'BPPDT', 'color': 'blue'},
               'RE_LAP_values_1': {'label': 'LAP', 'color': 'orange'}}

# 图像名称
text1 = "Obesity"
text2 = "Student Performance"
text3 = "Job Salary"
text4 = "OCD"

# 获取列名作为 x 轴数据
x_values = np.linspace(0, 1, len(data1_df.index))

# 设置字体大小
plt.rcParams.update({'font.size': 12})  # 根据需要调整字体大小

# 创建一个包含四个子图的图形
fig, axs = plt.subplots(1, 4, figsize=(18, 4))  # 共享y轴，以确保y轴标签位置一致

# 循环绘制每个子图
for i, data_df in enumerate([data1_df, data2_df, data3_df, data4_df]):
    # 获取每列的平均值作为 y 轴数据，并乘以 100
    for column in columns_to_plot:
        y_values = data_df[column] * 100

        # 绘制曲线
        axs[i].plot(x_values, y_values, label=column_info[column]['label'], color=column_info[column]['color'])

    # 设置 x 轴范围为 [0, 1]
    axs[i].set_xlim(0, 1)

    # 设置 x 轴标签
    axs[i].set_xlabel('B_prime')

    # 设置 y 轴标签（只在第一个子图上设置）
    if i == 0:
        axs[i].set_ylabel('RE(%)')
    
    # 在每张图像上方绘制名称
    if i == 0:
        axs[i].set_title(text1)
    elif i == 1:
        axs[i].set_title(text2)
    elif i == 2:
        axs[i].set_title(text3)
    elif i == 3:
        axs[i].set_title(text4)

# 创建空白的Line2D对象，用于模拟图例信息，仅在整体图上添加
lines = [Line2D([0], [0], color=column_info[column]['color'], label=column_info[column]['label'], linewidth=3) for column in columns_to_plot]

# 手动调整整体图的底部空白
plt.subplots_adjust(bottom=0.2)  # 可根据需要调整底部的空白

# 显示图例，增加整体图例的长度
axs[0].legend(handles=lines, loc='upper center', bbox_to_anchor=(0.6, -0.15), ncol=len(columns_to_plot), handlelength=5)  # 调整handlelength的值

# 保存图形
plt.savefig('1_mean.png')

# 显示图形
plt.show()
```
